```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
```

```python
SPReturn = pd.read_csv("SPreturn.csv")
n = SPReturn.shape[0]
```

```python
#Create a list of dates from 1980
dt = datetime.datetime(1980,1,1)
step = datetime.timedelta(days=1)
dates = []
for i in range(n):
    dates.append(dt.strftime("%y-%m-%d"))
    dt += step

dateindex = pd.bdate_range('1980-1-1', periods=n)
```

```python
SPReturn['date'] = dateindex
SPReturn.set_index('date')
```

```python
plt.plot(SPReturn['x'])
plt.ylabel('log return')
plt.xlabel('date')
plt.show()
```

```python
# Plot a histogram using 20 bins
R = SPReturn['x']
plt.hist(R, bins=40)
plt.xlabel('Return')
plt.ylabel('Number of Days Observed')
plt.title('Frequency Distribution of SP Returns');
plt.show()
```
